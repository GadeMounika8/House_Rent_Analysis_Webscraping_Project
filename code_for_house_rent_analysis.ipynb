```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
house =[]
for j in range(1,51):
    url = "https://www.makaan.com/hyderabad-residential-property/rent-property-in-hyderabad-city".format(j)
    page = requests.get(url)
    page = page.content
    soup = BeautifulSoup(page, 'html.parser')
    ol = soup.find('div')
    lists = ol.find_all('div', class_ = 'cardLayout clearfix')
    
    for list in lists:
        title = list.find('a', class_ = "typelink").text
        location = list.find('a', class_ = "loclink").text
        price = list.find('td', class_ = "price").text
        area = list.find('td', class_ = "size").text
        status = list.find('td', class_ = "val").text
        columns = [title, location, price, area, status]
        #print(columns)
        house.append([title, location, price, area, status])
```

```python
df = pd.DataFrame(house, columns = {'Title':title, 'Location':location, 'Price':price, 'Area':area, 'Status':status})
```
